```python
#evin m2 ile arsanın m2 ile topladım ve bahçenin toplam m2 buldum.
df["evin_bahçeyle_toplam_m2"] = df["sqft_living"] + df["sqft_lot"]
```

```python
#evin bodrum m2 ile evin m2 topladım 
df["evin_bodrumla_toplam_m2"] = df["sqft_above"] + df["sqft_basement"]
```

```python
# evin yapıldığı yıl tarihlerini yenilenme tarihi ile birleştirdim.
df['evin_yılı'] = df.apply(lambda row: row['yr_renovated'] if row['yr_renovated'] != 0 else row['yr_built'], axis=1)
```

```python
# tarih sütunundan sadece yıl bilgisini aldım.
df['year'] = df['date'].dt.year
```

```python
#evin yaşını bulmak için evin satış tarihinden evin yapıldığı yılı çıkradım
df["evin_yaşı"] = df["year"] - df["evin_yılı"]
```

```python
df = df.drop("lat", axis=1)
df = df.drop("long", axis=1)
df = df.drop("date", axis=1)
df = df.drop("id", axis=1)
df = df.drop("yr_built", axis=1)
df = df.drop("yr_renovated", axis=1)
df = df.drop("sqft_above", axis=1)
df = df.drop("sqft_basement", axis=1)
df = df.drop("sqft_living", axis=1)
df = df.drop("sqft_lot", axis=1)
df = df.drop("evin_yılı", axis=1)
```

```python
df.dtypes
```

```python
df['price'] = df['price'].astype(int)
df['bathrooms'] = df['bathrooms'].astype(int)
df['floors'] = df['floors'].astype(int)
```

```python
#veriler arasındaki ilişkiyi görmek için heatmap yaptım
corr_matrix=df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title('Korelasyon Isı Haritası')
plt.show()
```

### model oluşturma 

```python
# y=wx+b

y = df["price"].values

x= df.drop("price", axis=1).values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split ( x,y, test_size=0.20, random_state=42)
```

```python
len(x_train)
```

```python
#verilerin hepsinin belirli bir aralıkta olması için scale işlemi
from sklearn.preprocessing import MinMaxScaler
scaler= MinMaxScaler()
x_train = scaler.fit_transform(x_train)

x_test= scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model= Sequential()
```

```python
x_train.shape
```

```python
model.add(Dense(18, activation="relu"))
model.add(Dense(18, activation="relu"))
model.add(Dense(18, activation="relu"))
model.add(Dense(18, activation="relu"))
model.add(Dense(18, activation="relu"))

model.add(Dense(1))

model.compile(optimizer="adam", loss="mse")
```

```python
model.fit(x_train, y_train, validation_data=(x_test, y_test), batch_size=200, epochs=300)
```

```python
kayıpveri= pd.DataFrame(model.history.history)
```

```python
kayıpveri.plot()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
tahmin= model.predict(x_test)
mean_absolute_error(y_test, tahmin)
```
